```python
import numpy as np
from termcolor import colored

def flash_card(doc):
    questions = list(doc.keys())
    # Shuffle questions
    np.random.shuffle(questions)
    corrected_count = 0

    for num, q in enumerate(questions, 1):
        choices = doc[q][:-1]
        # Shuffle choices
        np.random.shuffle(choices)
        correct_pos = doc[q][-1]
        correct_answer = doc[q][correct_pos]

        # Print question and its number
        print(f'{num}. {q}')

        for index, choice in enumerate(choices, 1):
            print(f'({index}). {choice}')

        # Prompt the user to answer based on the number corresponding to the choices
        answer = int(input('What is the answer? '))
        correct = choices[answer - 1] == correct_answer

        if correct:
            corrected_count += 1
            print(correct)
        else:
            print("Wrong! Answer is ");
            print(colored(correct_answer, 'red'))
        print()
            
    print(f'Result: {corrected_count * 100 / (len(questions))}')
```
