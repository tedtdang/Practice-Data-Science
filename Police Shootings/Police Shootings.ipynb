```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

df = pd.read_csv('Police Shootings.csv')
```

```python
df.drop(columns='name', inplace=True)
```

```python
race_table = df.groupby('race').count().reset_index().sort_values('id').set_index('race')
```

```python
race_table.plot.pie(y='id', figsize=(5, 5))
```
