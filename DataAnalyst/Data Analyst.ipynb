# Import dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('DataAnalyst.csv')
df.drop(columns=['Company Name', 'Unnamed: 0'], inplace=True)
```

### Convert 'size' column to ordinal

```python
company_size = {
    '1 to 50 employees': 1,
    '51 to 200 employees': 2,
    '201 to 500 employees': 3,
    '501 to 1000 employees': 4,
    '1001 to 5000 employees': 5,
    '5001 to 10000 employees': 6,
    '10000+ employees': 7
}
df.Size = df.Size.map(company_size)
df.dropna(subset=['Size'], inplace=True)
```

### Turn 'Easy Apply' column to bool type

```python
df['Easy Apply'] = df['Easy Apply'].apply(lambda x: True if x == 'True' else False)
```

### Create 'salary_avg_estimate' column

```python
df.drop(df[df['Salary Estimate'] == '-1'].index, inplace = True)
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split()[0])
df['salary_min_estimate'] = df['Salary Estimate'].apply(lambda x: x.split('-')[0][1:-1]).astype(int)
df['salary_max_estimate'] = df['Salary Estimate'].apply(lambda x: x.split('-')[1][1:-1]).astype(int)
df['salary_avg_estimate'] = (df['salary_min_estimate'] + df['salary_max_estimate']) / 2
df.drop(columns='Salary Estimate', inplace=True)
```

### Convert 'Revenue' column to ordinal

```python
df.Revenue.unique()
```

```python
revenue_ordinal = {
    'Less than $1 million (USD)': 1,
    '$1 to $5 million (USD)': 2,
    '$5 to $10 million (USD)': 3,
    '$10 to $25 million (USD)': 4,
    '$25 to $50 million (USD)': 5,
    '$50 to $100 million (USD)': 6,
    '$100 to $500 million (USD)': 7,
    '$500 million to $1 billion (USD)': 8,
    '$1 to $2 billion (USD)': 9,
    '$2 to $5 billion (USD)': 10,
    '$5 to $10 billion (USD)': 11,
    '$10+ billion (USD)': 12,
    'Unknown / Non-Applicable': -1
}
df['revenue_size'] = df.Revenue.map(revenue_ordinal)
df.drop(columns='Revenue', inplace=True)
```

# Visualize the dataset

```python
sns.distplot(df['salary_avg_estimate'])
plt.title('Histogram of estimated average salary')
```

```python
plt.subplots(figsize=(15,15))
sns.boxplot(x='Rating', y='salary_avg_estimate', data=df)
plt.title('Barplot of company ratings vs estimated average salary')
```

```python
plt.subplots(figsize=(20,20))
plt.xticks(rotation=45)
sns.boxplot(x='Type of ownership', y='salary_avg_estimate', data=df)
plt.title('Barplot of type of ownership vs estimated average salary')
```

```python
plt.subplots(figsize=(20,20))
plt.xticks(rotation=45)
sns.boxplot(x='Sector', y='salary_avg_estimate', data=df)
plt.title('Barplot of company sector vs estimated average salary')
```
