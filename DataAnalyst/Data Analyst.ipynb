```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('DataAnalyst.csv')
df.drop(columns=['Company Name', 'Unnamed: 0'], inplace=True)
```

```python
df
```

### Convert size to ordinal

```python
company_size = {
    '1 to 50 employees': 1,
    '51 to 200 employees': 2,
    '201 to 500 employees': 3,
    '501 to 1000 employees': 4,
    '1001 to 5000 employees': 5,
    '5001 to 10000 employees': 6,
    '10000+ employees': 7
}
df.Size = df.Size.map(company_size)
df.dropna(subset=['Size'], inplace=True)
```

```python
df['Easy Apply'].unique()
```

### Handle Easy Apply

```python
df['Easy Apply'] = df['Easy Apply'].apply(lambda x: True if x == 'True' else False)
```
