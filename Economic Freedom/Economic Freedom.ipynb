```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_excel('Economic Freedom.xls')
df.drop(columns=['CountryID', 'Country Name', 'Country'], inplace=True)
```

```python
compared = df.drop(columns=['Region', 'Region Rank', 'World Rank', '2020 Score', 'Population (Millions)', 'GDP (Billions, PPP)'])
correlations = compared.corr(method ='pearson').unstack() 
```

```python
gdp_per_capita = correlations['GDP per Capita (PPP)']
strong = gdp_per_capita[gdp_per_capita.abs() > 0.5]
medium = gdp_per_capita[(gdp_per_capita.abs() > 0.2) & (gdp_per_capita.abs() < 0.5)]
weak = gdp_per_capita[gdp_per_capita.abs() <= 0.2]
print(strong, medium, weak, sep='\n\n')
```
