```python
# -*- coding: utf-8 -*-
"""
Created on Sun Sep 12 00:59:23 2021

@author: TeoHangXanh
"""
import random
import time
import math
from termcolor import colored

food_dict = {'Appetizer': ['Crispy Fried Pork Riblets',
                           'Crispy Egg Rolls',
                           'Rolls',
                           'Shrimp Blanket',
                           'Sweet Rice Chicken wings',
                           'Pork riblets',
                           'Lao Heavenly Beef',
                           'Lao Spicy Sausage',
                           'Nam Khao',
                           ],
             'Grill': ['Pork Skewer',
                       'Chicken Skewer',
                       'Beef Skewer',
                       'Marinated Beef Steak',
                       ],
             'Soup': ['Pho',
                      'Kao',
                      'Tom',
                      ],
             'Noodles': ['Pad Thai',
                         'Pad Sa Eew',
                         'Pad Kee Mow',
                         'Pad Woon Sen'
                         ],
             'Salads': ['Papaya Salad - Thum Ma Houng',
                        'Larb Beef - Lao Style',
                        'Larb Chicken - Lao Style'],
             'Curries': ['Green / Red / Yellow Curry'],
             'Entree': ['Sweet and Sour Chicken',
                        'Mongolian Beef',
                        'Mixed Vegies',
                        'Cashew Chicken',
                        'Pepper Steak',
                        'Thai Spicy Basil',
                        'Broccoli Beef',
                        ],
             'Fried Rice': ['Thai Fried Rice',
                            'Combination Fried Rice',
                            'Pineaple Fried Rice',
                            'Herbal Chicken Fried Rice',
                            'Crabmeat Fried Rice',
                            'Duck Fat Fried Rice',
                            ],
             'Sides': ['Sweet Rice',
                       'Jasmine Rice',
                       'Jeo Som',
                       'Jeo Bong',
                       'Jeo Mak Len',
                       ],
             'Kids': ['Egg Omelette - Lao Style',
                      'Chicken Nuggets',
                      'French Fries',
                      'Kao Chee',
                      ],
             'Senior': ['Thai Fried Rice with Mixed Vegies',
                        'Red-Yellow-Green Curry with Steamed Rice',
                        'Stir Fry Mixed Vegies with Steamed Rice',
                        'Sweet Rice Chicken Wings with Sweet Rice or Steamed',
                        ],
             }

lookup = {
    'A': 'Appetizer',
    'B': 'Grill',
    'C': 'Salads',
    'D': 'Soup',
    'E': 'Entree',
    'F': 'Sides',
    'G': 'Kids',
    'H': 'Senior',
    'I': 'Curries',
    'J': 'Fried Rice',
    'K': 'Noodles'
}
cat_and_name = []
for category, names in food_dict.items():
    for name in names:
        cat_and_name.append((category, name))
random.shuffle(cat_and_name)

scores = maximum_scores = 0
for food in cat_and_name:
    start = time.time()
    answer = lookup[input(
        f'What is the category of {food[1]}?\n{lookup}').strip().title()]
    if food[1] in food_dict[answer]:
        print(f'Correct!')
        scores += 1
    else:
        print(colored('Wrong!', 'red') + ' The category of ' +
              colored(food[1], 'yellow') + ' is ' + colored(food[0], 'magenta'))
    maximum_scores += 1
print(f'You got {scores / maximum_scores * 100}%')
```
