```python
import pickle
from textblob import TextBlob
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

with open("cleaned text.pkl", "rb") as f:
    tweets = pickle.load(f)
```

```python
# Store polarity and subjectivity of all tweets
polarity, subjectivity = [], []
for tweet in tweets:
    blob = TextBlob(tweet)
    polarity.append(blob.sentiment[0])
    subjectivity.append(blob.sentiment[1])
```

```python
def plot(data, title):
    '''Plot a list with mean and standard deviation'''
    plt.figure(figsize=(10, 7))
    sb.histplot(data, kde=True, bins=20).set(title=f'{title}. std = {np.std(data)}')
    plt.axvline(np.mean(data), color='r')
```

```python
# Plot polarity
plot(polarity, 'Polarity of tweets')
```

With the relatively symmetric bell shape, we can see that the majority of tweets are neutral 

```python
# Plot subjectivity
plot(subjectivity, 'Subjectivity of tweets')
```

We can see that the majority of tweets are skewed on fact (< 0.5).
