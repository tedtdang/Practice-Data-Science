### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least (difficult)

```python
def gender_to_numeric(x):
    if x == 'M':
        return 1
    return 0
users['gender_to_n'] = users.gender.apply(gender_to_numeric)
print(users.head())
a = users.groupby('occupation').gender_to_n.sum() * 100 / users.groupby('occupation').gender_to_n.count()
print(a.sort_values(ascending = False))
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.agg(['min', 'max'])
```

### Step 8.  For each occupation present the percentage of women and men (difficult)

```python
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
```
