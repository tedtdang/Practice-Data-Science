```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

df = pd.read_csv('meat-consumption-vs-gdp-per-capita.csv')
df['kg/person/year'] = df['Meat supply per person (kilograms per year)'] / df['Total population (Gapminder)']
```

```python
df.columns
```

```python
us = df[df.Entity == 'United States']
japan = df[df.Entity == 'Japan']
plt.figure(figsize=(15, 15)) 
plt.scatter(us.Year, us['Meat supply per person (kilograms per year)'])
plt.scatter(japan.Year, japan['Meat supply per person (kilograms per year)'])
plt.xlabel('Year')
plt.ylabel('Meat supply per person (kilograms per year)')
plt.title('Meat supply per person: US vs Japan')
plt.xticks(rotation='vertical')
plt.legend(['US', 'Japan'], loc ="lower right", fontsize=20)
```
