```python
# https://www.codewars.com/kata/591eab1d192fe0435e000014/train/python


def escape(carpark):
    def left(current_position, where_to_go):
        nonlocal direction
        direction.append("L" + str(current_position[1] - where_to_go[1]))
        current_position[1] = where_to_go[1]

    def right(current_position, where_to_go):
        nonlocal direction
        direction.append("R" + str(where_to_go[1] - current_position[1]))
        current_position[1] = where_to_go[1]

    def go_down(current_position):
        nonlocal direction
        if direction[-1][0] == "D":
            direction[-1] = "D" + str(int(direction[-1][-1]) + 1)
        else:
            direction.append("D1")
        current_position[0] += 1

    def ground(current_position):
        nonlocal carpark
        return current_position[0] == len(carpark)

    direction = []
    if carpark:
        pos = stairs = [-1, -1]
        exit = [len(carpark) - 1, len(carpark[0]) - 1]
        at_ground = False
        for i, v in enumerate(carpark):
            if 2 in v:
                pos = [i, v.index(2)]
                at_ground = pos[0] == exit[0]
                while not at_ground:
                    stairs = [pos[0], carpark[pos[0]].index(1)]
                    if pos[1] < stairs[1]:
                        right(pos, stairs)
                    elif pos[1] > stairs[1]:
                        left(pos, stairs)
                    else:
                        go_down(pos)
                        at_ground = pos[0] == exit[0]
                if at_ground:
                    if pos[1] < exit[1]:
                        right(pos, exit)
                    return direction
    return direction
```

```python
# https://www.codewars.com/kata/54f8693ea58bce689100065f/train/python

from math import gcd


def convert_fracts(lst):
    lcm = 1
    for i in lst:
        lcm = lcm * i[1] // gcd(lcm, i[1])
    return [[lcm // i[1] * i[0], lcm] for i in lst]
```

```python
# https://www.codewars.com/kata/53c93982689f84e321000d62/train/python

from functools import lru_cache


@lru_cache()
def is_prime(n):
    return all(n % d for d in range(3, int(n ** 0.5) + 1, 2)) or n == 2


@lru_cache()
def prime_list(n):
    """This returns a list from prime numbers from 2 to n"""
    if not isinstance(n, int):
        return False
    if n < 2:
        return []
    else:
        my_list = [2]
        for i in range(3, n + 1, 2):
            if is_prime(i):
                my_list.append(i)
    return my_list


@lru_cache()
def getAllPrimeFactors(n):
    answer = []
    if not isinstance(n, int):
        return answer
    if n <= 0:
        return []
    elif n == 1:
        return [1]
    while n > 1:
        for i in prime_list(n):
            if n % i == 0:
                n = n // i
                answer.append(i)
                break
    return answer


@lru_cache()
def getUniquePrimeFactorsWithCount(n):
    answer = [[], []]
    if not isinstance(n, int):
        return answer
    if n <= 0:
        return answer
    elif n == 1:
        return [[1], [1]]
    while n > 1:
        for i in prime_list(n):
            if n % i == 0:
                n = n // i
                if i in answer[0]:
                    i_index = answer[0].index(i)
                    answer[1][i_index] += 1
                else:
                    answer[0].append(i)
                    answer[1].append(1)
    return answer


@lru_cache()
def getUniquePrimeFactorsWithProducts(n):
    answer = []
    if not isinstance(n, int):
        return answer
    if n <= 0:
        return []
    return [i ** j for i, j in zip(*getUniquePrimeFactorsWithCount(n))]
```

```python
# https://www.codewars.com/kata/5603a9585480c94bd5000073/train/python
from functools import lru_cache


@lru_cache()
def sum_dif_rev(n):
    if n <= 0:
        return None
    num = 1
    count = 0
    while True:
        if num % 10 != 0:
            reversed_num = int(str(num)[::-1])
            plus = num + reversed_num
            minus = abs(num - reversed_num)
            if minus > 0 and plus % minus == 0:
                count += 1
            if count == n:
                return num
        num += 1
```

```python
# https://www.codewars.com/kata/515bb423de843ea99400000a/train/python


class PaginationHelper:
    # The constructor takes in an array of items and a integer indicating
    # how many items fit within a single page
    def __init__(self, collection, items_per_page):
        self.collection = collection
        self.items_per_page = items_per_page

    # returns the number of items within the entire collection
    def item_count(self):
        return len(self.collection)

    # returns the number of pages
    def page_count(self):
        pages = len(self.collection) / self.items_per_page
        return int(pages) if int(pages) == pages else int(pages) + 1

    # returns the number of items on the current page. page_index is zero based
    # this method should return -1 for page_index values that are out of range
    def page_item_count(self, page_index):
        a, b = divmod(len(self.collection), self.items_per_page)
        if a > page_index:
            return self.items_per_page
        elif a == page_index:
            return b
        else:
            return -1

    # determines what page an item is on. Zero based indexes.
    # this method should return -1 for item_index values that are out of range
    def page_index(self, item_index):
        if item_index < 0 or item_index >= len(self.collection):
            return -1
        else:
            return item_index // self.items_per_page
```

```python
# https://www.codewars.com/kata/532a69ee484b0e27120000b6/train/python


class Vector:
    def __init__(self, *args):
        if len(args) == 3:
            self.x, self.y, self.z = args
        else:
            self.x = args[0][0]
            self.y = args[0][1]
            self.z = args[0][2]

    def __add__(self, avec):
        return Vector(self.x + avec.x, self.y + avec.y, self.z + avec.z)

    def __sub__(self, avec):
        return Vector(self.x - avec.x, self.y - avec.y, self.z - avec.z)

    def __eq__(self, avec):
        return self.x == avec.x and self.y == avec.y and self.z == avec.z

    def cross(self, avec):
        return Vector(
            self.y * avec.z - self.z * avec.y,
            self.z * avec.x - self.x * avec.z,
            self.x * avec.y - self.y * avec.x,
        )

    def dot(self, avec):
        return self.x * avec.x + self.y * avec.y + self.z * avec.z

    def to_tuple(self):
        return (self.x, self.y, self.z)

    def __str__(self):
        return f"<{self.x}, {self.y}, {self.z}>"

    def __repre__(self):
        return f"<{self.x}, {self.y}, {self.z}>"

    @property
    def magnitude(self):
        return (self.x ** 2 + self.y ** 2 + self.z ** 2) ** 0.5
```

```python
# https://www.codewars.com/kata/572bbd7c72a38bd878000a73/solutions/python
import math


class Fraction:
    def __init__(self, numerator, denominator):
        self.top = numerator
        self.bottom = denominator

    def __eq__(self, other):
        first_num = self.top * other.bottom
        second_num = other.top * self.bottom
        return first_num == second_num

    def __add__(self, other):
        n = (self.top * other.bottom) + (other.top * self.bottom)
        d = self.bottom * other.bottom
        gcd = math.gcd(n, d)
        n /= gcd
        d /= gcd
        return Fraction(int(n), int(d))

    def __repr__(self):
        return f"{self.top}/{self.bottom}"
```

```python
# https://www.codewars.com/kata/587e1ef6f1a2534bbe0001ef/train/python

from dataclasses import dataclass


class DimensionsOutOfBoundError(Exception):
    def __init__(self, string, measured_value, upper):
        self.string = string
        self.measured_value = measured_value
        self.upper = upper

    def __str__(self):
        return f"Package {self.string}=={self.measured_value} out of bounds, should be: 0 < {self.string} <= {self.upper}"


@dataclass
class Package:
    length: int
    width: int
    height: int
    weight: int

    @property
    def length(self):
        return self._length

    @property
    def width(self):
        return self._width

    @property
    def height(self):
        return self._height

    @property
    def weight(self):
        return self._weight

    @property
    def volume(self):
        return self._length * self._width * self._height

    @length.setter
    def length(self, value):
        if value <= 0 or value > 350:
            raise DimensionsOutOfBoundError("length", value, 350)
        self._length = value

    @width.setter
    def width(self, value):
        if value <= 0 or value > 300:
            raise DimensionsOutOfBoundError("width", value, 300)
        self._width = value

    @height.setter
    def height(self, value):
        if value <= 0 or value > 150:
            raise DimensionsOutOfBoundError("height", value, 150)
        self._height = value

    @weight.setter
    def weight(self, value):
        if value <= 0 or value > 40:
            raise DimensionsOutOfBoundError("weight", value, 40)
        self._weight = value
```

```python
# https://www.codewars.com/kata/526d42b6526963598d0004db/train/python


class CaesarCipher(object):
    alphabet = "abcdefghijklmnopqrstuvwxyz"

    def __init__(self, shift):
        self.shift = shift

    def encode(self, st):
        return "".join(
            [
                CaesarCipher.alphabet[
                    (CaesarCipher.alphabet.index(i.lower()) + self.shift)
                    % len(CaesarCipher.alphabet)
                ].upper()
                if i.isalnum()
                else i
                for i in st
            ]
        )

    def decode(self, st):
        return "".join(
            [
                CaesarCipher.alphabet[
                    (CaesarCipher.alphabet.index(i.lower()) - self.shift)
                    % len(CaesarCipher.alphabet)
                ].upper()
                if i.isalnum()
                else i
                for i in st
            ]
        )
```

```python
# https://www.codewars.com/kata/52d1bd3694d26f8d6e0000d3/train/python


class VigenereCipher(object):
    gap = 3

    def __init__(self, key, alphabet):
        self.key = key
        self.alphabet = alphabet

    def encode(self, text):
        key = "".join(
            self.key[i % len(self.key)] if text[i].isalnum() else text[i]
            for i in range(len(text))
        )
        encoded = []

        for i in range(len(text)):
            if text[i] in self.alphabet:
                encoded.append(
                    self.alphabet[
                        (self.alphabet.index(key[i]) + self.alphabet.index(text[i]))
                        % len(self.alphabet)
                    ]
                )
            else:
                encoded.append(text[i])
        encoded = "".join(encoded)
        return encoded

    def decode(self, text):
        key = "".join(
            self.key[i % len(self.key)] if text[i].isalnum() else text[i]
            for i in range(len(text))
        )
        decoded = []

        for i in range(len(text)):
            if text[i] in self.alphabet:
                decoded.append(
                    self.alphabet[
                        (self.alphabet.index(text[i]) - self.alphabet.index(key[i]))
                        % len(self.alphabet)
                    ]
                )
            else:
                decoded.append(text[i])
        decoded = "".join(decoded)
        return decoded
```

```python
# https://www.codewars.com/kata/57520361f2dac757360018ce/train/python

from dataclasses import dataclass


@dataclass
class PokeScan:
    name: str
    level: int
    pkmntype: int

    def info(self):
        adj = {"water": "wet", "fire": "fiery", "grass": "grassy"}.get(
            self.pkmntype, None
        )

        power = {
            lambda x: x <= 20: "weak",
            lambda x: 20 < x <= 50: "fair",
            lambda x: x > 50: "strong",
        }
        power = "".join([value for key, value in power.items() if key(self.level)])
        return f"{self.name}, a {adj} and {power} Pokemon."
```

```python
# https://www.codewars.com/kata/586c0909c1923fdb89002031/train/python

import numpy as np


class Connect4:
    current_player = 2
    stadium = np.zeros((6, 7))
    game_over = False

    def __init__(self, new_instance=None):
        if new_instance is None:
            Connect4.current_player = 2
            Connect4.stadium = np.zeros((6, 7))
            Connect4.game_over = False

    @staticmethod
    def same(arr):
        """This returns True if the array has a same value for all elements"""
        return np.all(arr == arr[0])

    def is_win(self):
        """This return True if there are 4 consecutive points either horizontally, vertically, or diagonally"""
        for i in range(len(Connect4.stadium)):
            for j in range(len(Connect4.stadium[i])):
                if Connect4.stadium[i, j] > 0:
                    # Check horizontally
                    if j < 4:
                        if self.same(Connect4.stadium[i, j : j + 4]):
                            return True
                    # Check vertically
                    if i < 3:
                        if self.same(Connect4.stadium[i : i + 4, j]):
                            return True
                    # Check diagonally
                    # Check top left
                    if (
                        i > 2
                        and j > 2
                        and Connect4.stadium[i, j]
                        == Connect4.stadium[i - 1][j - 1]
                        == Connect4.stadium[i - 2][j - 2]
                        == Connect4.stadium[i - 3][j - 3]
                    ):
                        return True
                    # Check top right
                    if (
                        i > 2
                        and j < 4
                        and Connect4.stadium[i, j]
                        == Connect4.stadium[i - 1][j + 1]
                        == Connect4.stadium[i - 2][j + 2]
                        == Connect4.stadium[i - 3][j + 3]
                    ):
                        return True
                    # Check bottom left
                    if (
                        i < 3
                        and j > 2
                        and Connect4.stadium[i, j]
                        == Connect4.stadium[i + 1][j - 1]
                        == Connect4.stadium[i + 2][j - 2]
                        == Connect4.stadium[i + 3][j - 3]
                    ):
                        return True
                    # Check bottom right
                    if (
                        i < 3
                        and j < 4
                        and Connect4.stadium[i, j]
                        == Connect4.stadium[i + 1][j + 1]
                        == Connect4.stadium[i + 2][j + 2]
                        == Connect4.stadium[i + 3][j + 3]
                    ):
                        return True
        return False

    def play(self, col):
        if Connect4.game_over:
            return "Game has finished!"

        if np.all(Connect4.stadium[:, col] > 0):
            return f"Column full!"
        else:
            Connect4.current_player = 1 if Connect4.current_player == 2 else 2
            i = 5
            while i > 0 and Connect4.stadium[i, col]:
                i -= 1
            if Connect4.stadium[i, col] == 0:
                Connect4.stadium[i, col] = Connect4.current_player
        if self.is_win():
            Connect4.game_over = True
            return f"Player {Connect4.current_player} wins!"
        return f"Player {Connect4.current_player} has a turn"
```

```python
# https://www.codewars.com/kata/587ac5616d360f6bed000088/train/python

from dataclasses import dataclass
import numpy as np


@dataclass
class Grid:
    width: int
    height: int

    def __init__(self, x, y):
        self.matrix = [[0 for _ in range(y)] for _ in range(x)]

    @property
    def grid(self):
        return "".join(
            str(j) for i in self.matrix for j in "".join(str(num) for num in i) + "\n"
        )[:-1]

    def plot_point(self, x, y):
        self.matrix[y - 1][x - 1] = "X"
        return self.matrix
```
