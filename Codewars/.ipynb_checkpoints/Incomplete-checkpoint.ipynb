```python
# https://www.codewars.com/kata/59c191df4f98a8a70b00001e/train/python

class HexCipher:
    @staticmethod
    def encode(s, n=0):
        if n == 0:
            return s

        def encode(s):
            return "".join([hex(ord(i))[-2:] for i in s])

        s = encode(s)
        for _ in range(n - 1):
            s = encode(s)
        return s

    @staticmethod
    def decode(s, n=0):
        if n == 0:
            return s

        def decode(s):
            return "".join(
                [chr(int(hex(int(s[i : i + 2], 16)), 16)) for i in range(0, len(s), 2)]
            )

        s = decode(s)
        for _ in range(n - 1):
            s = decode(s)
        return s
```

```python
# https://www.codewars.com/kata/5941c545f5c394fef900000c/train/python


class Warrior:
    rank = (
        "Pushover",
        "Novice",
        "Fighter",
        "Warrior",
        "Veteran",
        "Sage",
        "Elite",
        "Conqueror",
        "Champion",
        "Master",
        "Greatest",
    )

    def __init__(self):
        self.experience = 100
        self.rank = "Pushover"
        self.achievements = []

    @property
    def level(self):
        return self._level

    @level.setter
    def level(self):
        self._level = self.experience // 100 if self.experience < 100_000 else 100

    @property
    def rank(self):
        return self._rank

    @rank.setter
    def rank(self):
        if self._rank != "Greatest":
            rank_index = Warrior.rank.index(self.level // 10)
            self._rank = Warrio.rank[rank_index]

    @property
    def experience(self):
        return self._experience

    @experience.setter
    def experience(self):
        self._experience = 100

    @experience.setter
    def experience(self, value):
        self._experience += value

    def training(self, train):
        # Check enemy level
        if train[2] < 1 or train[2] > 100:
            return "Not strong enough"
        else:
            self.experience += train[1]
            self.level_up
            self.rank_up
            self.achievements.append(train[0])
            return train[0]

    def battle(self, level):
        # Retrieve rank from level
        rank_index = Warrior.rank.index(level // 10)
        rank = Warrio.rank[rank_index]
        if level < 1 or level > 100:
            return "Invalid level"
        if self.rank <= rank - 1 and self.level <= level - 5:
            return "You've been defeated"
        else:
            if self.level > level + 1:
                return "Easy fight"
            elif self.level == level + 1:
                self.experience += 5
                return "A good fight"
            elif self.level == level:
                self.experience += 10
                return "A good fight"
            elif self.level + 1 == level:
                diff = level - self.level
                self.experience += diff * diff * 20
                return "An intense fight"
            else:
                self.experience += 10
                return "An intense fight"
```

```python
# https://www.codewars.com/kata/55cd4ce59382498cbd000080/train/python

class List:
    def __init__(self, type):
        self.type = type
        self.count = 0

    def add(self, item):
        if type(item) != self.type:
            return f"This item is not of type: {self.type.__name__}"
        self.count += 1
        return item
```

```python
# https://www.codewars.com/kata/5b09cd44ee196bf290000082/train/python

def direction(arr, pos):
    if pos[0] % 2 == 0:  # to the right
        if pos[1] < len(arr[pos[0]]) - 1:
            return [pos[0], pos[1] + 1]
        elif pos[1] == len(arr[pos[0]]) - 1:
            return [pos[0] + 1, pos[1]]
    else:
        if pos[1] > 0:
            return [pos[0], pos[1] - 1]
        else:
            return [pos[0] + 1, pos[1]]


def after_strike(monster, damage):
    initial_damage = damage
    death_toll = 0
    while damage > 0:
        if monster["hitpoints"] < damage:
            monster["number"] -= 1
            death_toll += 1
        damage -= monster["hitpoints"]
    damage = initial_damage
    if death_toll == 1:
        return f"{death_toll} {monster['type']}"
    elif death_toll > 1:
        return f"{death_toll} {monster['type']}s"


def chain_lightning(monster_arr, coordinates, spellpower):
    report = []
    sub_list = []
    damage = 25 * spellpower * 2
    current_pos = coordinates
    while current_pos[0] < len(monster_arr) and current_pos[1] < len(
        monster_arr[current_pos[0]]
    ):
        monster = monster_arr[current_pos[0]][current_pos[1]]
        if monster is None:
            damage = damage // 2
            current_pos = direction(monster_arr, current_pos)
            if sub_list is not None:
                report.append(sub_list)
                sub_list = []
                print(f"sub_list = {sub_list}, report = {report}")
            continue
        else:
            if monster["resistance"] < 100:
                damage = damage // 2
            if monster is not None and sub_list is not None:
                sub_list.append(after_strike(monster, damage))
            print(f"sub_list = {sub_list}, report = {report}")
        current_pos = direction(monster_arr, current_pos)
    report = report.append(filter(None, sub_list))
    return report
```

```python
# https://www.codewars.com/kata/5e8dd197c122f6001a8637ca

def de_dup(a, b):
    a = [i for i in a if i not in b]
    c = []
    for i in b:
        if i not in c:
            c.append(i)
    return a, c


def remove_duplicate_ids(obj):
    print(obj)
    for k1, v1 in obj.items():
        for k2, v2 in obj.items():
            if k1 != k2 and set(v1) & set(v2) != set():
                if int(k1) <= int(k2):
                    a, b = v1, v2
                    obj[k1], obj[k2] = de_dup(a, b)
                else:
                    a, b = v2, v1
                    obj[k2], obj[k1] = de_dup(a, b)
    return obj
```

```python
# https://www.codewars.com/kata/586669a8442e3fc307000048/train/python

from math import pi


class Shape:
    def __init__(self, side):
        self.side = side
        self.area = self.side

    def __lt__(self, other):
        return self.area < other.area

    def __le__(self, other):
        return self.area <= other.area


class Square(Shape):
    def __init__(self, side):
        super().__init__(side)

    def area(self):
        return self.side ** 2


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height / 2


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return self.radius * pi


class CustomShape(Shape):
    def __init__(self, area):
        self.area = area
```
