```python
# https://www.codewars.com/kata/59c191df4f98a8a70b00001e/train/python

class HexCipher:
    @staticmethod
    def encode(s, n=0):
        if n == 0:
            return s

        def encode(s):
            return "".join([hex(ord(i))[-2:] for i in s])

        s = encode(s)
        for _ in range(n - 1):
            s = encode(s)
        return s

    @staticmethod
    def decode(s, n=0):
        if n == 0:
            return s

        def decode(s):
            return "".join(
                [chr(int(hex(int(s[i : i + 2], 16)), 16)) for i in range(0, len(s), 2)]
            )

        s = decode(s)
        for _ in range(n - 1):
            s = decode(s)
        return s
```

```python
# https://www.codewars.com/kata/5941c545f5c394fef900000c/train/python


class Warrior:
    rank = (
        "Pushover",
        "Novice",
        "Fighter",
        "Warrior",
        "Veteran",
        "Sage",
        "Elite",
        "Conqueror",
        "Champion",
        "Master",
        "Greatest",
    )

    def __init__(self):
        self.experience = 100
        self.rank = "Pushover"
        self.achievements = []

    @property
    def level(self):
        return self._level

    @level.setter
    def level(self):
        self._level = self.experience // 100 if self.experience < 100_000 else 100

    @property
    def rank(self):
        return self._rank

    @rank.setter
    def rank(self):
        if self._rank != "Greatest":
            rank_index = Warrior.rank.index(self.level // 10)
            self._rank = Warrio.rank[rank_index]

    @property
    def experience(self):
        return self._experience

    @experience.setter
    def experience(self):
        self._experience = 100

    @experience.setter
    def experience(self, value):
        self._experience += value

    def training(self, train):
        # Check enemy level
        if train[2] < 1 or train[2] > 100:
            return "Not strong enough"
        else:
            self.experience += train[1]
            self.level_up
            self.rank_up
            self.achievements.append(train[0])
            return train[0]

    def battle(self, level):
        # Retrieve rank from level
        rank_index = Warrior.rank.index(level // 10)
        rank = Warrio.rank[rank_index]
        if level < 1 or level > 100:
            return "Invalid level"
        if self.rank <= rank - 1 and self.level <= level - 5:
            return "You've been defeated"
        else:
            if self.level > level + 1:
                return "Easy fight"
            elif self.level == level + 1:
                self.experience += 5
                return "A good fight"
            elif self.level == level:
                self.experience += 10
                return "A good fight"
            elif self.level + 1 == level:
                diff = level - self.level
                self.experience += diff * diff * 20
                return "An intense fight"
            else:
                self.experience += 10
                return "An intense fight"
```

```python
# https://www.codewars.com/kata/55cd4ce59382498cbd000080/train/python

class List:
    def __init__(self, type):
        self.type = type
        self.count = 0

    def add(self, item):
        if type(item) != self.type:
            return f"This item is not of type: {self.type.__name__}"
        self.count += 1
        return item
```
