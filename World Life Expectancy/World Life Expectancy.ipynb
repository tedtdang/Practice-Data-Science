```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('World Life Expectancy.csv')
```

### Find the best n_clusters

```python
from sklearn.cluster import KMeans
no_country = df.drop(columns='Countryandregions')
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(no_country)
    sse[k] = kmeans.inertia_ # SSE to closest cluster centroid
plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
top20_both = df.sort_values(by='Bothsexesrank')[:20]
top20_male = df.sort_values(by='Malerank')[:20]
top20_female = df.sort_values(by='Femalerank')[:20]
```

```python
fig = plt.gcf()
fig.set_size_inches(20, 10)
ax = sns.barplot(x='Countryandregions', y='Bothsexeslifeexpectancy', data=top20_both)
ax.set(xlabel='Countries', ylabel="Both sexes' life expectancy", title = "Top 20's life expectancy of both sexes")
plt.xticks(rotation=45)
```
